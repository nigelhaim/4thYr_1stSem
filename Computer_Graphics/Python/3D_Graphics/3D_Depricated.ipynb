```python
#DRAWING A CONE
# Parameters for the cone
radius = 3  # Cone radius at the base
height = 6  # Height of the cone
num_points = 200  # Number of points around the cone
```

```python
# Create a meshgrid for the cone's angle (theta) and height (z)
theta = np.linspace(0, 2*np.pi, num_points)
z = np.linspace(0, height, num_points)
theta, Z = np.meshgrid(theta, z)
```

```python
# Parametric equations for the cone
X = radius * (1 - Z/height) * np.cos(theta)
Y = radius * (1 - Z/height) * np.sin(theta)
```

```python
# Plot the cone in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=5, cstride=5, color='RED', edgecolor='BLUE')

# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Display the plot
plt.show()
```

```python
#DRAWING A SPHERE
# Define the radius of the sphere
radius = 4
num_points = 200
```

```python
# Create a meshgrid for the spherical coordinates (theta, phi)
theta = np.linspace(0, 2 * np.pi, num_points)  # Azimuthal angle
phi = np.linspace(0, np.pi, num_points)        # Polar angle
theta, phi = np.meshgrid(theta, phi)
```

```python

# Parametric equations for the sphere
x = radius * np.sin(phi) * np.cos(theta)
y = radius * np.sin(phi) * np.sin(theta)
z = radius * np.cos(phi)
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the sphere
ax.plot_surface(x, y, z, color='RED', edgecolor='BLUE', linewidth=0.5)

# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Display the plot
plt.show()
```

```python
def distance_3d(point1, point2):
  """Calculates the Euclidean distance between two points in 3D space.

  Args:
    point1: A tuple or list representing the coordinates of the first point (x1, y1, z1).
    point2: A tuple or list representing the coordinates of the second point (x2, y2, z2).

  Returns:
    The Euclidean distance between the two points.
  """
  # Convert the points to NumPy arrays for efficient calculations
  point1 = np.array(point1)
  point2 = np.array(point2)

  # Calculate the distance using the formula: sqrt((x2 - x1)^2 + (y2 - y1)^2 + (z2 - z1)^2)
  distance = np.sqrt(np.sum((point2 - point1)**2))

  return distance

# Example usage:
point_a = (1, 2, 3)
point_b = (4, 5, 6)

distance = distance_3d(point_a, point_b)
print(f"The distance between {point_a} and {point_b} is: {distance}")
```

```python
#DRAWING PYRAMID
# Parameters for the pyramid
base_size = 16  # Side length of the square base
height = 9     # Height of the pyramid

# Define the vertices of the pyramid
vertices = np.array([
    [-base_size/2, -base_size/2, 0],  # Base corner 1
    [ base_size/2, -base_size/2, 0],  # Base corner 2
    [ base_size/2,  base_size/2, 0],  # Base corner 3
    [-base_size/2,  base_size/2, 0],  # Base corner 4
    [ 0, 0, height]                   # Apex
])

# Define the faces of the pyramid (4 triangles + 1 square base)
faces = [
    [vertices[0], vertices[1], vertices[4]],  # Front face
    [vertices[1], vertices[2], vertices[4]],  # Right face
    [vertices[2], vertices[3], vertices[4]],  # Back face
    [vertices[3], vertices[0], vertices[4]],  # Left face
    [vertices[0], vertices[1], vertices[2], vertices[3]]  # Base face
]
```

```python
# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the pyramid
ax.add_collection3d(Poly3DCollection(faces, facecolors='RED', linewidths=1, edgecolors='BLUE', alpha=.5))

# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the limits for the axes
ax.set_xlim([-base_size, base_size])
ax.set_ylim([-base_size, base_size])
ax.set_zlim([0, height])

# Display the plot
plt.show()
```

```python
#DRAWING VOXELS
# Parameters for the voxel grid
grid_size = 5  # Size of the grid (5x5x5 grid of voxels)
voxel_size = 1  # Size of each voxel (cube)
```

```python
# Create a figure and 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Loop over the 3D grid and draw cubes (voxels)
for x in range(grid_size):
    for y in range(grid_size):
        for z in range(grid_size):
            # Voxel's corner coordinates
            x_pos = x * voxel_size
            y_pos = y * voxel_size
            z_pos = z * voxel_size

            # Create a voxel as a 3D cube using plot_surface
            # Each voxel is represented by a box defined by its corners
            ax.bar3d(x_pos, y_pos, z_pos, voxel_size, voxel_size, voxel_size, color='YELLOW', linewidths=0.5)

# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the viewing angle
ax.view_init(20, 75)

# Show the plot
plt.show()
```

```python
#Union of Sphere and Cube
def create_sphere(radius=0.8):
       theta = np.linspace(0, 2 * np.pi, 100)
       phi = np.linspace(0, np.pi, 100)
       theta, phi = np.meshgrid(theta, phi)
       x = radius * np.sin(phi) * np.cos(theta)
       y = radius * np.sin(phi) * np.sin(theta)
       z = radius * np.cos(phi)
       return x, y, z

def create_cube(size=1):
       vertices = np.array([
           [0, 0, 0], [size, 0, 0], [size, size, 0], [0, size, 0],
           [0, 0, size], [size, 0, size], [size, size, size], [0, size, size]
       ])
       faces = [
           [vertices[0], vertices[1], vertices[2], vertices[3]],
           [vertices[4], vertices[5], vertices[6], vertices[7]],
           [vertices[0], vertices[1], vertices[5], vertices[4]],
           [vertices[2], vertices[3], vertices[7], vertices[6]],
           [vertices[1], vertices[2], vertices[6], vertices[5]],
           [vertices[4], vertices[7], vertices[3], vertices[0]]
       ]
       return vertices, faces
```

```python
sphere_x, sphere_y, sphere_z = create_sphere()
cube_vertices, cube_faces = create_cube()
```
