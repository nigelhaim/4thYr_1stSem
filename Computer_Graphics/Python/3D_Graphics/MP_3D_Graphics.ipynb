1. Create a 3D line plot that shows the path of a sine wave in 3D space.


```python
amplitude = 1
frequency = 2 * np.pi
phase_shift = 0
```

```python
# function to draw a line 

line_x = np.linspace(0, 5*np.pi, num=100)
line_y = np.linspace(0, 5*np.pi, num=100)
X, Y = np.meshgrid(line_x, line_y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
#plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=5, cstride=5, color='green', edgecolor='ORANGE')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

2. Find the distance of the following: 
    - a. X to Y 
    - b. X to Z 
    - where X(2, 3, 6), Y(7, 8, 9), and Z (2, 9, 4).

a. compute x to y

```python
def distance_3d(point1, point2):
  point1 = np.array(point1)
  point2 = np.array(point2)

  distance = np.sqrt(np.sum((point2 - point1)**2))

  return distance
```

```python
# Compute x to y 

x = (2, 3, 6)
y = (7, 8, 9)

d = distance_3d(x, y)
d
```

```python
print("The distance of x to y is:",float(round(d, 2)))
```

b. b. X to Z 

```python
# Compute x to z 

x = (2, 3, 6)
z = (2, 9, 4)

d = distance_3d(x, z)
d
```

```python
print("The distance of x to z is:",float(round(d, 2)))
```
