## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import math
```

### a. Cylinder whose radius = 13.3, height = 17 and number of points/resolution = 450 (yellow)

```python
#DRAWING A CYLINDER
# Parameters for the cylinder
radius = 13.3  # Cylinder radius
height = 17  # Cylinder height
num_points = 450  # Number of points around the cylinder's circumference

# Create a meshgrid for the cylinder's angle (theta) and height (z)
theta = np.linspace(0, 2*np.pi, num_points)  # Angle around the cylinder
z = np.linspace(0, height, num_points)      # Height of the cylinder
theta, Z = np.meshgrid(theta, z)

# Parametric equations for a cylinder
X = radius * np.cos(theta)
Y = radius * np.sin(theta)
```

```python
# Plot the cylinder
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=5, cstride=5, color='yellow', edgecolor='RED')
# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# Show the plot
plt.show()
```

```python
def volume_cylinder(radius, height):
    return math.pi * radius**2 * height

radius_cylinder = radius
height_cylinder = height

# Calculating volumes
print("Volume of Cylinder:", volume_cylinder(radius_cylinder, height_cylinder))
```

### b. Cone whose radius = 13.3, height = 17 and number of points/resolution = 450 (cyan)

```python
#DRAWING A CONE
# Parameters for the cone
radius = 13.3  # Cone radius at the base
height = 17  # Height of the cone
num_points = 450  # Number of points around the cone
# Create a meshgrid for the cone's angle (theta) and height (z)
theta = np.linspace(0, 2*np.pi, num_points)
z = np.linspace(0, height, num_points)
theta, Z = np.meshgrid(theta, z)
# Parametric equations for the cone
X = radius * (1 - Z/height) * np.cos(theta)
Y = radius * (1 - Z/height) * np.sin(theta)
```

```python
# Plot the cone in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=5, cstride=5, color='cyan', edgecolor='RED')

# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Display the plot
plt.show()
```

```python
def volume_cone(radius, height):
    return (1/3) * math.pi * radius**2 * height

radius_cone = radius
height_cone = height

print("Volume of Cone:", volume_cone(radius_cone, height_cone), "cubic units")
```

### c. Sphere whose radius = 13.3 and number of points/resolution = 450 (magenta)

```python
#DRAWING A SPHERE
# Define the radius of the sphere
radius = 13.3
num_points = 450

# Create a meshgrid for the spherical coordinates (theta, phi)
theta = np.linspace(0, 2 * np.pi, num_points)  # Azimuthal angle
phi = np.linspace(0, np.pi, num_points)        # Polar angle
theta, phi = np.meshgrid(theta, phi)


# Parametric equations for the sphere
x = radius * np.sin(phi) * np.cos(theta)
y = radius * np.sin(phi) * np.sin(theta)
z = radius * np.cos(phi)
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the sphere
ax.plot_surface(x, y, z, color='magenta', edgecolor='RED', linewidth=0.5)

# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Display the plot
plt.show()
```

```python

def volume_sphere(radius):
    return (4/3) * math.pi * radius**3

radius_sphere = radius

# Calculating volumes
print("Volume of Sphere:", volume_sphere(radius_sphere), 'cubic units')
```

### d. Cube whose initial coordinate located at (4, 8, 9) and the edge is 12. (blue)

```python
initial_corner = np.array([4, 8, 9])
edge_length = 12

# Calculate the vertices of the cube
vertices = [
    initial_corner, 
    initial_corner + [edge_length, 0, 0], 
    initial_corner + [edge_length, edge_length, 0], 
    initial_corner + [0, edge_length, 0],
    initial_corner + [0, 0, edge_length], 
    initial_corner + [edge_length, 0, edge_length], 
    initial_corner + [edge_length, edge_length, edge_length], 
    initial_corner + [0, edge_length, edge_length], 
]

# Define the faces of the cube using vertex indices
faces = [
    [vertices[0], vertices[1], vertices[2], vertices[3]],  # Bottom
    [vertices[4], vertices[5], vertices[6], vertices[7]],  # Top
    [vertices[0], vertices[1], vertices[5], vertices[4]],  # Front
    [vertices[2], vertices[3], vertices[7], vertices[6]],  # Back
    [vertices[1], vertices[2], vertices[6], vertices[5]],  # Right
    [vertices[4], vertices[7], vertices[3], vertices[0]]   # Left
]
```

```python
# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Add the faces to the plot
ax.add_collection3d(Poly3DCollection(faces, facecolors='blue', linewidths=1, edgecolors='RED', alpha=.30))

# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set limits for the axes to encompass the cube
ax.set_xlim([initial_corner[0], initial_corner[0] + edge_length])
ax.set_ylim([initial_corner[1], initial_corner[1] + edge_length])
ax.set_zlim([initial_corner[2], initial_corner[2] + edge_length])

# Show the plot
plt.show()
```

```python
def distance_3d(point1, point2):
  """Calculates the Euclidean distance between two points in 3D space.

  Args:
    point1: A tuple or list representing the coordinates of the first point (x1, y1, z1).
    point2: A tuple or list representing the coordinates of the second point (x2, y2, z2).

  Returns:
    The Euclidean distance between the two points.
  """
  # Convert the points to NumPy arrays for efficient calculations
  point1 = np.array(point1)
  point2 = np.array(point2)

  # Calculate the distance using the formula: sqrt((x2 - x1)^2 + (y2 - y1)^2 + (z2 - z1)^2)
  distance = np.sqrt(np.sum((point2 - point1)**2))

  return distance

# Example usage:
point_a = (4, 8, 9)
point_b = (16, 9, 8)

distance = distance_3d(point_a, point_b)
print(f"The distance between {point_a} and {point_b} is: {distance}")
```

```python
def volume_cube(side_cube):
    return side_cube**3
side_cube = int(distance)

# Calculating volumes
print("Volume of Cube:", volume_cube(side_cube), 'cubic units')
```

### e. Pyramid whose base is rectangle (L = 21 and W = 16) and height is 23 units. (green)

```python
#DRAWING PRISM
# Parameters for the rectangular prism
length = 21  # Length of the prism
width = 16  # Width of the prism
height =23  # Height of the prism

# Define the vertices of the rectangular prism
vertices = np.array([
    [-length/2, -width/2, 0],  # Base corner 1
    [ length/2, -width/2, 0],  # Base corner 2
    [ length/2,  width/2, 0],  # Base corner 3
    [-length/2,  width/2, 0],  # Base corner 4
    [ 0, 0, height]    
])

faces = [
    [vertices[0], vertices[1], vertices[4]],  # Front face
    [vertices[1], vertices[2], vertices[4]],  # Right face
    [vertices[2], vertices[3], vertices[4]],  # Back face
    [vertices[3], vertices[0], vertices[4]],  # Left face
    [vertices[0], vertices[1], vertices[2], vertices[3]]  # Base face
]
```

```python
translation = [5,7,2]
```

```python
# Apply translation
x_translated = x + translation[0]
y_translated = y + translation[1]
z_translated = z + translation[2]

# Plot original and translated spheres
fig = plt.figure(figsize=(10, 5))

# Original Sphere
ax1 = fig.add_subplot(121, projection='3d')
ax.add_collection3d(Poly3DCollection(faces, facecolors='green', linewidths=1, edgecolors='RED', alpha=.5))
ax1.set_title('Original Sphere')
ax1.set_xlim([0, 100])
ax1.set_ylim([00, 100])
ax1.set_zlim([00, 100])

# Translated Sphere
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(x_translated, y_translated, z_translated, color='red', alpha=0.7)
ax2.set_title('Translated Sphere')
ax2.set_xlim([0, 100])
ax2.set_ylim([0, 100])
ax2.set_zlim([0, 100])

plt.show()
```
