## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import math
```

### 1. Translate the prism in the last activity using the parameter [5, 7, 2]

```python
translation = [5,7,2]
```

```python
#DRAWING PRISM
# Parameters for the rectangular prism
length = 21  # Length of the prism
width = 16   # Width of the prism
height = 23  # Height of the prism

# Define the vertices of the rectangular prism
vertices = np.array([
    [0, 0, 0],  # Vertex 0
    [length, 0, 0],  # Vertex 1
    [length, width, 0],  # Vertex 2
    [0, width, 0],  # Vertex 3
    [0, 0, height],  # Vertex 4
    [length, 0, height],  # Vertex 5
    [length, width, height],  # Vertex 6
    [0, width, height],  # Vertex 7
])

translated_vertices = vertices + translation

# Define the 6 faces of the rectangular prism
faces = [
    [vertices[0], vertices[1], vertices[2], vertices[3]],  # Bottom face
    [vertices[4], vertices[5], vertices[6], vertices[7]],  # Top face
    [vertices[0], vertices[1], vertices[5], vertices[4]],  # Front face
    [vertices[1], vertices[2], vertices[6], vertices[5]],  # Right face
    [vertices[2], vertices[3], vertices[7], vertices[6]],  # Back face
    [vertices[3], vertices[0], vertices[4], vertices[7]],  # Left face
]

translated_faces = [
    [translated_vertices[0], translated_vertices[1], translated_vertices[2], translated_vertices[3]],  # Bottom face (translate z)
    [translated_vertices[4], translated_vertices[5], translated_vertices[6], translated_vertices[7]],  # Top face (translate z)
    [translated_vertices[0], translated_vertices[1], translated_vertices[5], translated_vertices[4]],  # Front  (translate x)
    [translated_vertices[1], translated_vertices[2], translated_vertices[6], translated_vertices[5]],  # Right face (translate y)
    [translated_vertices[2], translated_vertices[3], translated_vertices[7], translated_vertices[6]],  # Back face (translate x)
    [translated_vertices[3], translated_vertices[0], translated_vertices[4], translated_vertices[7]],  # Left face (translate y)
]
```

```python
# Apply translation
# x_translated = faces + translation[0]
# y_translated = faces + translation[1]
# z_translated = faces + translation[2]

# Plot original and translated spheres
fig = plt.figure(figsize=(10, 5))

# Original Prism
ax1 = fig.add_subplot(121, projection='3d')
ax1.add_collection3d(Poly3DCollection(faces, facecolors='orange', linewidths=1, edgecolors='RED', alpha=.5))
ax1.set_title('Original Prism')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_xlim([0, length+10])
ax1.set_ylim([0, width+10])
ax1.set_zlim([0, height+10])

# Translated Sphere
ax2 = fig.add_subplot(122, projection='3d')
ax2.add_collection3d(Poly3DCollection(translated_faces, facecolors='orange', linewidths=1, edgecolors='RED', alpha=.5))
ax2.set_xlim([0, length+10])
ax2.set_ylim([0, width+10])
ax2.set_zlim([0, height+10])
plt.show()
```

### 2. Scale the cube in the last activity using the parameter [0, 2, 1]

```python
scaling_factors = [0,2,1]
```

```python
initial_corner = np.array([4, 8, 9])
edge_length = 12

# Calculate the vertices of the cube
vertices = [
    initial_corner, 
    initial_corner + [edge_length, 0, 0], 
    initial_corner + [edge_length, edge_length, 0], 
    initial_corner + [0, edge_length, 0],
    initial_corner + [0, 0, edge_length], 
    initial_corner + [edge_length, 0, edge_length], 
    initial_corner + [edge_length, edge_length, edge_length], 
    initial_corner + [0, edge_length, edge_length], 
]

scaled_vertices = np.array(vertices) * scaling_factors

# Define the faces of the cube using vertex indices
faces = [
    [vertices[0], vertices[1], vertices[2], vertices[3]],  # Bottom
    [vertices[4], vertices[5], vertices[6], vertices[7]],  # Top
    [vertices[0], vertices[1], vertices[5], vertices[4]],  # Front
    [vertices[2], vertices[3], vertices[7], vertices[6]],  # Back
    [vertices[1], vertices[2], vertices[6], vertices[5]],  # Right
    [vertices[4], vertices[7], vertices[3], vertices[0]]   # Left
]

scaled_faces = [
    [scaled_vertices[0], scaled_vertices[1], scaled_vertices[2], scaled_vertices[3]],  # Bottom
    [scaled_vertices[4], scaled_vertices[5], scaled_vertices[6], scaled_vertices[7]],  # Top
    [scaled_vertices[0], scaled_vertices[1], scaled_vertices[5], scaled_vertices[4]],  # Front
    [scaled_vertices[2], scaled_vertices[3], scaled_vertices[7], scaled_vertices[6]],  # Back
    [scaled_vertices[1], scaled_vertices[2], scaled_vertices[6], scaled_vertices[5]],  # Right
    [scaled_vertices[4], scaled_vertices[7], scaled_vertices[3], scaled_vertices[0]]   # Left
]
```

```python
# Plot original and scaled cylinders
fig = plt.figure(figsize=(10, 5))

# Original Cylinder
ax1 = fig.add_subplot(121, projection='3d')
ax1.add_collection3d(Poly3DCollection(faces, facecolors='blue', linewidths=1, edgecolors='RED', alpha=.30))

# Set labels for clarity
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

ax1.set_title('Original Cube')
ax1.set_xlim([initial_corner[0]-10, initial_corner[0] + edge_length+10])
ax1.set_ylim([initial_corner[1]-10, initial_corner[1] + edge_length+10])
ax1.set_zlim([initial_corner[2]-10, initial_corner[2] + edge_length+10])


# Scaled Cylinder
ax2 = fig.add_subplot(122, projection='3d')
ax2.add_collection3d(Poly3DCollection(scaled_faces, facecolors='blue', linewidths=1, edgecolors='RED', alpha=.30))
ax2.set_title('Scaled Cube')
ax2.set_xlim([initial_corner[0]-10, initial_corner[0] + edge_length+10])
ax2.set_ylim([initial_corner[1]-10, initial_corner[1] + edge_length+10])
ax2.set_zlim([initial_corner[2]-10, initial_corner[2] + edge_length+10])

plt.show()
```
